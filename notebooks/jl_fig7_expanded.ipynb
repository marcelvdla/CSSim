```julia
# number of systems 
n = 10

# number of rows and columns product must == n
nrows = 5
ncols = 2

# Constant distance
D = 900
rand_dists = D*rand(n-1)
@show rand_dists
params = Dict{Symbol,Any}(
        :ρ  => 4.2, 
        :f  => 1.0,
        :α₀ => -1.0, 
        :w₀ =>  1.0,
        :a₁ => 1.0, 
        :h  => 2.0, 
        :a₂ => 0.0, 
        :d  => repeat([D / (n-1)], n-1), 
        :l  => 600.0, 
        :P₀ => 1.0, 
        :β₁ => 0.0, 
        :β₂ => 1.0,
        :n  => n,
        :ecosystems_to_deforest => [],
    )

# plot the systems
fig, axs = phase_portrait(
    50,
    params;
    nrows=nrows,
    ncols=ncols,
    figsize=(20, 12),
    Δt = 0.005,
    sharex=false,
)

fig.suptitle(
    "Complex 10-Forest System Phase Portraits without Perturbation", 
    fontsize=32)

fig
```
